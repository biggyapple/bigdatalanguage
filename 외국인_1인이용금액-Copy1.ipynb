```python
my_list=['쇼핑','숙박','식음료','소매','문화/레져']
for i in range(len(my_list)):
    condition = (chinese.업종 == my_list[i])
    chinese = chinese.sort_values(by=['업종','1인당 이용금액'],ascending=[True,False])
    a=chinese[condition].head()
    b=a["읍면동명"]
    c=a["1인당 이용금액"]
    plot_a=pd.concat([b,c], axis=1)
    plot_chinese=plot_a.set_index('읍면동명')
    plot_chinese
    plot_chinese.plot(kind='bar')
    plt.title(my_list[i], fontsize='20')
```
